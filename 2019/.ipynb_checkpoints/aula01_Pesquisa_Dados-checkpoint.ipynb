É uma ciência baseada na **Teoria da Probabilidade**, cujo objetivo principal é nos auxiliar a tomar decisões ou tirar conclusões em situações de *incerteza*, a partir de dados.

### População:
É uma coleção completa de todos os elementos a serem estudados e que possuem certa característica em comum. A coleção é completa quando inclui todos os sujeitos a serem estudados. A população pode ser *finita* ou *infinita*.

    - *População alvo*: 
    conjunto de elementos que queremos abranger em nosso estudo. São os elementos para os quais desejamos que as conclusões oriundas da pesquisa sejam válidas.
    
    - *População acessível*: 
    conjunto de elementos que queremos abranger em nosso estudo e que são passíveis de serem observados, com respeito às características que pretendemos levantar.

### Censo:
É o conjunto de dados obtidos de ***TODOS*** os membros da população

### Amostra:
Seleção de uma parte da população a ser observada (plano de amostragem para garantir representabilidade)

### Variável:
Qualquer característica de interesse associada aos elementos de uma população. Classificação segundo à natureza:
    - Qualitativo Nominal
    - Qualitativo Ordinal
    - Quantitativo Discreto
    - Quantitativo Contínuo

### Dado:
São observações coletadas (por exemplo, medidas, sexo, resposta de um questionário, etc.) dos elementos de estudo.
A coleta de dados representa um dos aspectos da Estatística.

# COLETA DE DADOS

## Planejamento de coleta de dados

- **Dados secundários**: já existem em alguma publicação ou referência. Podem reduzir os custos da pesquisa.
- **Dados primários**: dados observados diretamente
- A ética é um ponto importante na elaboração da coleta de dados:
    - Dados amostrais devem ser coletados de modo apropriado, usando um processo de seleção aleatória
    - Dados coletados de modo não apropriado podem ser inúteis a análise estatística
    - Existem diferentes técnicas de amostragem

## Amostragem
### Amostragem por conveniência
Coletam-se elementos de fácil acesso ou de interesse para o estudo.

### Amostragem Probabilística
Cada elemento da população tem uma chance de ser selecionado.

   - *Amostragem por conglomerado*: 
    Divide-se a população em blocos (homogêneos) e amostram-se os blocos aleatoriamente e todos os indivíduos dentro do bloco são entrevistados.
    
   - *Amostragem estratificada*: 
    divide-se a população em grupo segundo algum estrato (pelo menos dois), amostram-se aleatoriamente indivíduos dentro de cada grupo.
    
   - *Amostragem sistemática*: 
    selecionar um elemento da populaçõa a cada *k* elementos.
    
   - *Amostragem aleatória simples*: 
    cada elemento tem a mesma chance de ser amostrado.

## Banco de dados

```python
from sklearn.datasets import load_wine
data = load_wine()
data
```

```python
from pandas import DataFrame
```

```python
df = DataFrame(data.data, columns=data.feature_names)
df
```

```python
df.plot.scatter("alcohol","magnesium")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(DataFrame(df['alcohol']),DataFrame(df['magnesium']))
```

```python
reg.coef_[0][0]
```

```python
reg.intercept_[0]
```

```python
reg.score(DataFrame(df['alcohol']),DataFrame(df['magnesium']))
```
