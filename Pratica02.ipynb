```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('_data/weight-height.csv')
df.head()
```

```python
df.describe()
```

```python
df['Peso [kg]'] = df['Weight'] *.453592
df['Altura [m]'] = df['Height'] *.0254
df['IMC'] = df['Peso [kg]'] / (df['Altura [m]'] * df['Altura [m]'])

df.head()
```

```python
def imcCategoria(value):
    if(value<18.5):
        return 'Magreza'
    elif(value<24.9):
        return 'Saudável'
    elif(value<29.9):
        return 'Sobrepeso'
    elif(value<34.9):
        return 'Obesidade G1'
    elif(value<39.9):
        return 'Obesidade Severa G2'
    else:
        return 'Obesidade Mórbida G3'
```

```python
df['IMC cat'] = df['IMC'].apply(imcCategoria)
df.head()
```
